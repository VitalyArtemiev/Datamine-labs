```python
import matplotlib.pyplot as plt

import numpy as np

stDev = 100
m = np.random.normal(loc = 1, scale = stDev, size = (1000,50))
print(m)

(h, e) = np.histogram(np.ndarray.flatten(m), bins = 100)
print(h)

plt.hist(m, bins = 10)  
plt.title("Initial histogram")
plt.show()

plt.hist(np.ndarray.flatten(m), bins = 100)  
plt.title("Initial histogram")
plt.show()

m = (m - np.mean(m, 1, keepdims = True)) / np.std(m, 1, keepdims = True)

(h, e) = np.histogram(np.ndarray.flatten(m), bins = 100)
print(h)

plt.hist(m, bins = 10)  
plt.title("Histogram normalized")
plt.show()

plt.hist(np.ndarray.flatten(m), bins = 100)  
plt.title("Histogram normalized")
plt.show()
```

```python
m = np.random.randint(0, 6, size = (5,5))
r = []

for i, row in enumerate(m):
    if row.sum() > 10:
        r.append(i)
        
print(m)
print(r)
```
